# 3. Calculate Angles

```python
def calculate_angle(a,b,c):
    a = np.array(a) 
    b = np.array(b) 
    c = np.array(c) 
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 
```

```python
def master(img):
    with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:

        image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        results = pose.process(image)
    
        # Recolor back to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        landmarks = results.pose_landmarks.landmark
        ## Left Hand
        shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
        elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
        wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
        angle_left_hand = calculate_angle(shoulder, elbow, wrist)
        ## Right Hand
        shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
        elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
        wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]
        angle_Right_hand = calculate_angle(shoulder, elbow, wrist)
        ## Left SHoulder
        x1 = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
        x2 = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
        x3 = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
        angle_Right_shoulder = calculate_angle(x1, x2, x3)
        ## Left SHoulder
        x1 = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
        x2 = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
        x3 = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
        angle_Left_shoulder = calculate_angle(x1, x2, x3)

        ## Right knee
        x1 = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
        x2 = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
        x3 = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
        angle_Right_knee = calculate_angle(x1, x2, x3)
        ## Left knee
        x1 = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
        x2 = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
        x3 = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
        angle_Left_knee = calculate_angle(x1, x2, x3)
        ## Head
        x1 = [landmarks[mp_pose.PoseLandmark.NOSE.value].x,landmarks[mp_pose.PoseLandmark.NOSE.value].y]
        x2 = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
        x3 = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
        Head = calculate_angle(x1, x2, x3)
        # Kamar
        x1 = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
        x2 = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
        x3 = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
        kamar = calculate_angle(x1, x2, x3)
        return Head, angle_left_hand, angle_Right_hand, angle_Right_shoulder, angle_Left_shoulder, angle_Right_knee, angle_Left_knee, kamar

master_worrier = cv2.imread(r'D:\University\Computer Vision\Project\worrier_pic.PNG') 
worrier_M_Head, worrier_M_angle_left_hand, worrier_M_angle_Right_hand, worrier_M_angle_Right_shoulder, worrier_M_angle_Left_shoulder, worrier_M_angle_Right_knee, worrier_M_angle_Left_knee, worrier_M_kamar = master(master_worrier)
master_downdog = cv2.imread(r'D:\University\Computer Vision\Project\downdog.PNG') 
dog_m_Head, dog_m_angle_left_hand, _, _, dog_m_angle_Left_shoulder, _, dog_m_angle_Left_knee, dog_m_kamar = master(master_downdog)




def worrier(M_H, M_L_h, M_R_h, M_R_s, M_L_S, M_R_k, M_L_k, M_k ,img,L_shoulder,L_elbow,R_shoulder,R_elbow,Head,kamar,R_knee, L_knee):
    H, L_h, R_h, R_s, L_S, R_k, L_k, k = master(img)
    if abs(M_H-H) > 20:
        if M_H > H:
            cv2.putText(img,"open", 
                           tuple(np.multiply(Head, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
        else:
            cv2.putText(img,"Close", 
                           tuple(np.multiply(Head, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )


    if abs(M_L_h-L_h) > 20:
        if M_L_h > L_h:
            cv2.putText(img,"open", 
                           tuple(np.multiply(L_elbow, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
        else:
            cv2.putText(img,"Close", 
                           tuple(np.multiply(L_elbow, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )


    if abs(M_R_h-R_h) > 20:
        if M_R_h > R_h:
            cv2.putText(img,"open", 
                           tuple(np.multiply(R_elbow, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
        else:
            cv2.putText(img,"Close", 
                           tuple(np.multiply(R_elbow, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )


    if abs(M_R_s-R_s) > 20:
        if M_R_s > R_s:
            cv2.putText(img,"open", 
                           tuple(np.multiply(R_shoulder, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
        else:
            cv2.putText(img,"Close", 
                           tuple(np.multiply(R_shoulder, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )


    if abs(M_L_S-L_S) > 20:
        if M_L_S > L_S:
            cv2.putText(img,"open", 
                           tuple(np.multiply(L_shoulder, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
        else:
            cv2.putText(img,"Close", 
                           tuple(np.multiply(L_shoulder, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )


    if abs(M_R_k-R_k) > 20:
        if M_R_k > R_k:
            cv2.putText(img,"open", 
                           tuple(np.multiply(R_knee, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
        else:
            cv2.putText(img,"Close", 
                           tuple(np.multiply(R_knee, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )


    if abs(M_L_k-L_k) > 20:
        if M_L_k > L_k:
            cv2.putText(img,"open", 
                           tuple(np.multiply(L_knee, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
        else:
            cv2.putText(img,"Close", 
                           tuple(np.multiply(L_knee, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )


    if abs(M_k-k) > 20:
        if M_k > k:
            cv2.putText(img,"open", 
                           tuple(np.multiply(kamar, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
        else:
            cv2.putText(img,"Close", 
                           tuple(np.multiply(kamar, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )

            
def downdog(M_H, M_L_h, M_L_S, M_L_k, M_k ,img,L_shoulder,L_elbow,R_shoulder,R_elbow,Head,kamar,R_knee, L_knee):
    H, L_h, _, _, L_S, _, L_k, k = master(img)
    if abs(M_H-H) > 20:
        if M_H > H:
            cv2.putText(img,"open", 
                           tuple(np.multiply(Head, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
        else:
            cv2.putText(img,"Close", 
                           tuple(np.multiply(Head, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )


    if abs(M_L_h-L_h) > 20:
        if M_L_h > L_h:
            cv2.putText(img,"open", 
                           tuple(np.multiply(L_elbow, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
        else:
            cv2.putText(img,"Close", 
                           tuple(np.multiply(L_elbow, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )


    if abs(M_L_S-L_S) > 20:
        if M_L_S > L_S:
            cv2.putText(img,"open", 
                           tuple(np.multiply(L_shoulder, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
        else:
            cv2.putText(img,"Close", 
                           tuple(np.multiply(L_shoulder, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )


    if abs(M_L_k-L_k) > 20:
        if M_L_k > L_k:
            cv2.putText(img,"open", 
                           tuple(np.multiply(L_knee, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
        else:
            cv2.putText(img,"Close", 
                           tuple(np.multiply(L_knee, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )


    if abs(M_k-k) > 20:
        if M_k > k:
            cv2.putText(img,"open", 
                           tuple(np.multiply(kamar, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
        else:
            cv2.putText(img,"Close", 
                           tuple(np.multiply(kamar, [img.shape[1], img.shape[0]]).astype(int)), 
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                                )
```

```python
import cv2
import time
import numpy as np
import joblib
import pickle
import copy
import tensorflow as tf

model = tf.keras.models.load_model(r'D:\University\Computer Vision\Project\Xception.h5')
cap = cv2.VideoCapture(0)

frame_count = 0
start_time = time.time()

frame = None
S_frame = None 
kernel = np.ones((5,5), np.uint8)


with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if S_frame is None:
            S_frame = frame
            continue
        f_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        f_blurred = cv2.GaussianBlur(f_gray, (21,21), 0)
        s_gray = cv2.cvtColor(S_frame, cv2.COLOR_BGR2GRAY)
        s_blurred = cv2.GaussianBlur(s_gray, (21,21), 0)
        delta = cv2.absdiff(s_blurred, f_blurred)
        thresh = cv2.threshold(delta, 107, 255, cv2.THRESH_BINARY)[1]
        thresh = cv2.dilate(thresh, kernel, iterations=15)

        mode = cv2.RETR_EXTERNAL
        method = cv2.CHAIN_APPROX_SIMPLE
        
        img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (229,229), interpolation = cv2.INTER_AREA)
        img = img.reshape(1,229,229,3)/255.0
        prediction = model.predict(img)
        prediction=np.argmax(prediction)
        text_list = ['downdog', 'goddess', 'plank', 'tree', 'warrior2']
        if prediction == 0:
            text= text_list[0]
        if prediction == 1:
            text= text_list[1]
        elif prediction == 2:
            text= text_list[2]
        elif prediction == 3:
            text= text_list[3]
        elif prediction == 4:
            text= text_list[4]
        
        cv2.putText(frame, str(text), (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2,cv2.LINE_AA)



        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
      
        results = pose.process(image)
    
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        try:
            landmarks = results.pose_landmarks.landmark
            
            L_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            L_elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            R_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
            R_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
            R_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
            L_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
            Head = [landmarks[mp_pose.PoseLandmark.NOSE.value].x,landmarks[mp_pose.PoseLandmark.NOSE.value].y]
            kamar = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            if text == 'warrior2':
                worrier(worrier_M_Head, worrier_M_angle_left_hand, worrier_M_angle_Right_hand, worrier_M_angle_Right_shoulder, worrier_M_angle_Left_shoulder, worrier_M_angle_Right_knee, worrier_M_angle_Left_knee, worrier_M_kamar ,frame,L_shoulder,L_elbow,R_shoulder,R_elbow,Head,kamar,R_knee, L_knee)
            if text == 'downdog':
                downdog(dog_m_Head, dog_m_angle_left_hand, dog_m_angle_Left_shoulder, dog_m_angle_Left_knee, dog_m_kamar ,frame,L_shoulder,L_elbow,R_shoulder,R_elbow,Head,kamar,R_knee, L_knee)

                       
        except:
            pass
        imgs = frame.copy()
        mp_drawing.draw_landmarks(imgs, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2), 
                                mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2) 
                                 )           
        cv2.imshow('img2', imgs)    
        cnts , _ = cv2.findContours(thresh.copy(), mode, method)
        if len(cnts)>0:
            for c in cnts:
                if cv2.contourArea(c)<100:
                    continue
                (x,y,w,h) = cv2.boundingRect(c)
                #cv2.putText(frame, 'Vibration Alert!', (x,y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2,cv2.LINE_AA)
                #cv2.rectangle(frame, (x,y), (x+w, y+h), (0,0,255), 3)

        

        frame_count += 1
        end_time = time.time()
        elapsed_time = end_time - start_time
        fps = frame_count / elapsed_time
        cv2.putText(frame, 'FPS :'+str(fps), (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.imshow('img', frame)
        s_frame = frame
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```
